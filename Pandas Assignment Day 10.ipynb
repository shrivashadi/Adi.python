## Groupby

```python
# Create a dataframe using dictionary of your choice

data = data = pd.DataFrame({
    'dept': ['HR', 'IT', 'IT', 'HR', 'Finance', 'Finance'],
    'name': ['Amit', 'Raj', 'Priya', 'Neha', 'Suresh', 'Kavya'],
    'salary': [50000, 80000, 75000, 55000, 60000, 65000]
})
print(data)


```

```python
# Use Groupby of single column with aggregate sum()

print(data.groupby('dept')['salary'].sum())
```

```python
# Use Groupby of single column with aggregate count()
# Use Groupby of single column with aggregate count()
print(data.groupby('dept')['salary'].count())


```

```python
# Use Groupby of single column with aggregate min() and max()
print(data.groupby('dept')['salary'].min())
print(data.groupby('dept')['salary'].max())
```

```python

# Use Groupby of any 2 columns with aggregate mean()
print(data.groupby(['dept', 'name'])['salary'].mean())

```

```python
# Use Groupby of any 2 columns with aggregate min() and max()
print(data.groupby(['dept', 'name'])['salary'].agg(['min', 'max']))

```
